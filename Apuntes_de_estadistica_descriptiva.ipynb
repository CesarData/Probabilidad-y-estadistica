## **Introducción** <a name="introduccion"></a>

La estadística es una de las áreas de las matemáticas de mayor relevancia para la ciencia de datos, sin ella no sería posible. Para conocer más de ella vamos a comenzar por definir algunos conceptos centrales tomados principalmente de Mario F. Triola (2018) quien define la **estadística** como “la ciencia que se encarga de planear estudios y experimentos, obtener datos y luego organizar, resumir, presentar, analizar e interpretar esos datos para obtener conclusiones basadas en ellos.” (pág. 4). En otras palabras, la estadística es una ciencia que procesa datos, desde su recolección hasta su interpretación. <br><br>

Aquí surge el concepto de **datos** el cual debemos entender como un conjunto de observaciones, es decir, la recolección o registro de los atributos que nos interesan de los elementos de estudio. Estos elementos pueden ser personas, cosas o eventos y la totalidad de ellos conforman la **población** de nuestro objeto de estudio o investigación. En ocasiones será posible estudiar la totalidad de la población. Pensemos, por ejemplo, en los estudiantes activos de una universidad. La institución tendrá registros de todos sus alumnos y podrá analizar datos relacionados con ellos, como la edad, promedio, licenciatura que estudian, etc. Cuando nos encontramos en esta situación donde todos los datos requeridos de la población están disponibles se tiene un **censo** (Devore, 2012, pág. 3) y cada medición de estas características es un **parámetro**.  Sin embargo, lo más común es que no sea factible tener acceso a todos los miembros de la población, ya sea por tiempo, costo, disponibilidad, entre otras razones. Así que trabajamos con un subconjunto de la población que sea representativo (e idealmente aleatorio) llamada **muestra** y cada medición de sus características es llamada **dato estadístico** (Triola, 2018, pág. 13) los cuales son utilizados para estudiar a la población de la que provienen para describirla o hacer inferencias. Así que podemos hablar de dos ramas de la estadística. Una para describir y resumir las características más significativas de los datos estudiados conocida como **estadística descriptiva** mientras que la **estadística inferencial** es como el pensamiento inductivo ya que pretende inferir generalizaciones de la población a partir de una muestra (Triola, 2018, pág. 81).


## **Tipos de datos** <a name="datos"></a>

Los datos estudiados son características que cambian de un elemento a otro por ello se les conoce como variables (Devore, 2012, pág. 3) y son de diferentes tipos:<br><br>


- **Datos de tipo atributo**, o **cualitativos**: Corresponden a observaciones sobre cualidades de un objeto o individuo. Suelen codificarse por medio de palabras, pero también se pueden usar números que jueguen el papel de etiquetas. Los datos cualitativos son aquellos que pueden ser iguales o diferentes, pero que no admiten ningún otro tipo de comparación significativa.
Es decir, que no tenga ningún sentido preguntarse si uno es más grande que otro, ni efectuar operaciones aritméticas con ellos, aunque estén representados por números. En el contexto de los data frames son llamados **factores**.


- **Datos ordinales**: Similares a los cualitativos, con la única diferencia de que se pueden ordenar de manera natural. Por ejemplo, las calificaciones en un control (suspenso, aprobado, notable, sobresaliente).

- **Datos cuantitativos**: Se refieren a medidas, tales como edades, longitudes, etc, son los que expresan cantidades que se representan mediante números. Éstos se suelen clasificar en continuos y discretos.

  - Los **datos continuos** son los que, si existiese la posibilidad de medirlos con precisión infinita, en principio podrían tomar todos los valores de un intervalo de la recta real. A modo de ejemplo, el peso, la altura, el tiempo... son datos de este tipo.

  - Por su parte, los **datos discretos** son los que pueden tomar un solo conjunto contable de valores. El número de colores de un gato, el número de individuos que conforman una población son algunos ejemplos de este tipo de datos. Conviene tener en cuenta que esta división es solo teórica. Es decir, en la práctica, todos estos datos son discretos puesto que la precisión infinita no existe. Sin embargo, es necesario de vez en cuando suponer los datos de tipo continuo para así poder utilizar técnicas específicas en su análisis.<br><br>

## **Estudio de frecuencias** <a name="frecuencias"></a>

Una tabla de distribución de frecuencias nos permite organizar los datos. La tabla comúnmente presenta cuatro columnas aunque pueden ser más o menos según la necesidad del estudio.<br><br>

- La **frecuencia absoluta**, $n_j$, del nivel $l_j$ en esta variable cualitativa es el número de observaciones en las que $x_i$ toma el valor $l_j$.<br><br>

- La **frecuencia relativa** del nivel $l_j$ en esta variable cualitativa es la fracción<br><br>
$$f_j = \frac{n_j}{n}$$
<br><br>
Es decir, la frecuencia relativa del nivel $l_j$ es la fracción (en tanto por uno) de observaciones que corresponden a este nivel. <br><br>

- La **frecuencia absoluta acumulada** es contar cuantas veces hemos observado un dato menor o igual a $x_i$.<br><br>

- La **frecuencia relativa acumulada** es la fracción del total de las observaciones en tanto por 1 que representa su frecuencia absoluta acumulada.

```python
def freq_table(data, column):
  tablaFrecuencias = data.groupby(column).agg(frequency=(column, "count"))
  tablaFrecuencias["Frecuencia acumulada"] = tablaFrecuencias["frequency"].cumsum()
  tablaFrecuencias["Frecuencia relativa"] = tablaFrecuencias["frequency"] / tablaFrecuencias["frequency"].sum()
  tablaFrecuencias["Porcentaje de descubrimiento"] = round(tablaFrecuencias["Frecuencia relativa"] * 100, 4).astype(str)+' %'
  tablaFrecuencias["Frecuencia relativa acumulada"] = tablaFrecuencias["frequency"].cumsum() / tablaFrecuencias["frequency"].sum()
  tablaFrecuencias["Porcentaje acumulado"] = round(tablaFrecuencias["Frecuencia relativa acumulada"] * 100, 4).astype(str)+' %'
  tablaFrecuencias = tablaFrecuencias.rename(columns={'frequency': 'Frecuencia absoluta'})
  return tablaFrecuencias
```

## **Medidas de tendencia central** <a name="central"></a>

Las medidas de tendencia central son las que dan un valor representativo del centro de todas las observaciones. Algunas de las más importantes son:<br><br>

- La **media aritmética** o **valor medio**<br><br>
$$\bar{x} = \frac{\sum_{i=1}^nx_i}{n}=\frac{\sum_{j=1}^kn_jX_j}{n}=\sum_{j=1}^kf_jX_j$$<br><br>

- La **mediana**, que representa el valor central en la lista ordenada de observaciones. Es menos susceptible a los valores atípicos, contrario a la media que se ve facílmente afectada por ellos. La definición formal de la mediana es la siguiente. Denotando por<br><br>

 $$x_{(1)}\le x_{(2)}\le\dots\le x_{(n)}$$<br><br>

  los datos de la variable cuantitativa ordenados de menor a mayor, la mediana es

  - Si $n$ par, la medio de los dos datos centrales $$\frac{x_{(\frac{n}{2})}+x_{(\frac{n}{2}+1)}}{2}$$<br><br>

  - Si $n$ impar, el dato central $x_{(\frac{n+1}{2})}$<br><br>


- La **moda** es el valor (o valores) de máxima frecuencia (absoluta o relativa, el resultado será el mismo).<br><br>

- El **mitad de rango**, que es la diferencia entre el máximo y el mínimo de las observaciones entre 2.<br><br>

$$mitad de rango=\frac{max - min}{2}$$

## **Medidas de posición** <a name="posicion"></a>

Dada una proporción $p\in(0,1)$, el **cuantil de orden $p$** de una variable cuantitativa, $Q_p$, es el valor más pequeño tal que su frecuencia relativa acumulada es mayor o igual a $p$.<br><br>

Dicho de otro modo, si tenemos un conjunto de observaciones $x_1,\dots,x_n$ y los ordenamos de menor a mayor, entonces $Q_p$ será el número más pequeño que deja a su izquierda (incluyéndose a sí mismo) como mínimo a la fracción $p$ de los datos. Es decir, $p\cdot n$.<br><br>

Así, ahora es más claro ver que la mediana vendría a ser $Q_{0.5}$, el cuantil de orden 0.5.<br><br>

Algunos cuantiles tienen nombre propio:<br><br>

- Los **cuartiles** son los cuantiles $Q_{0.25},Q_{0.5}$ y $Q_{0.75}$. Respectivamente, son llamados primer, segundo y tercer cuartil. El primer cuartil, $Q_{0.25}$, será el menor valor que es mayor o igual a una cuarta parte de las observaciones y $Q_{0.75}$, el menor valor que es mayor o igual a tres cuartas partes de los datos observados.
- El cuantil $Q_{0.5}$ es la **mediana**.
- Los **deciles** son los cuantiles $Q_p$ con $p$ un múltiplo de 0.1.
- Los **percentiles** son son los cuantiles $Q_p$ con $p$ un múltiplo de 0.01.<br><br>

**Cuantiles**<br><br>

La definición de cuantil anteriormente dada es orientativa. La realidad es que, exceptuando el caso de la mediana, no hay consenso sobre cómo deben calcularse los cuantiles. En verdad, existen diferentes métodos que pueden dar lugar a soluciones distintas.<br><br>

Al fin y al cabo, nuestro objetivo no es el de encontrar el primer valor de una muestra cuya frecuencia relativa acumulada en la variable sea mayor o igual a $p$, sino estimar el valor de esta cantidad para el total de la población.<br><br>

Para calcular los cuantiles de orden $p$ de una variable cualitativa $x$ con R, se utiliza la instrucción `quantile(x,p)`, la cual dispone de 9 métodos diferentes que se especifican con el parámetro `type`. El valor por defecto es `type = 7` y no hace falta especificarlo.<br><br>

El **rango intercuartílico**, que es la diferencia entre el tercer y primer cuartil, $Q_{0.75}-Q_{0.25}$.


## **Medidas de dispersión** <a name="dispersion"></a>

Las **medidas de dispersión** evalúan lo dispersos que están los datos. Algunas de las más importantes son:<br><br>

- El **rango** o **recorrido**, que es la diferencia entre el máximo y el mínimo de las observaciones.<br><br>

$$rango=max - min$$<br><br>

- La **varianza**, a la que denotaremos por $s^2$, es la media aritmética de las diferencias al cuadrado entre los datos $x_i$ y la media aritmética de las observaciones, $\bar{x}$. <br><br>

$$s^2 = \frac{\sum_{j=1}^n(x_j-\bar{x})^2}{n}=\frac{\sum_{j=1}^kn_j(X_j-\bar{x})^2}{n}=\sum_{j=1}^kf_j(X_j-\bar{x})^2$$.<br><br>

- La **desviación típica** es la raíz cuadrada positiva de la varianza, $s=\sqrt{s^2}$.<br><br>

- La **varianza muestral** es la corrección de la varianza. La denotamos por $\tilde{s}^2$ y se corresponde con<br><br>

$$\tilde{s}^2 = \frac{n}{n-1}s^2 = \frac{\sum_{j=1}^n(x_i-\bar{x})^2}{n-1}$$<br><br>

- La **desviación típica muestral**, que es la raíz cuadrada positiva de la varianza muestral, $\tilde{s} = \sqrt{\tilde{s}^2}$<br><br>

### **Propiedades de la varianza**<br><br>


- $s^2\ge 0$. Esto se debe a que, por definición, es una suma de cuadrados de números reales.<br><br>

- $s^2 = 0\Longrightarrow x_j-\bar{x}=0\ \forall j= 1,\dots,n$. En consecuencia, si $s^2=0$, entonces todos los datos son iguales.<br><br>

- $s^2 =\frac{\sum_{j=1}^nx_j^2}{n}-\bar{x}^2$. Es decir, la varianza es la media de los cuadrados de los datos menos el cuadrado de la media aritmética de estos.<br><br>

### **Varianza y varianza muestral**<br><br>

La diferencia entre ambas definiciones viene por la interrelación entre la estadística descriptiva y la inferencial.

Por un lado, es normal medir cómo varían los datos  cuantitativos mediante su varianza definida como la media aritmética de las distancias al cuadrado de los datos a su valor medio. No obstante, por otro lado, el conjunto de nuestras observaciones, por lo normal, será una muestra de una población mucho mayor y nos interesará estimar entre otras muchas cosas su variabilidad.

La varianza de una muestra suele dar valores más pequños que la varianza de la población, mientras que la varianza muestral tiende a dar valores alrededor de la varianza de la población.<br><br>

### **Varianza y varianza muestral**<br><br>

Esta corrección, para el caso de una muestra grande no es notable. Dividir $n$ entre $n-1$ en el caso de $n$ ser grande no significa una gran diferencia y aún menos si tenemos en cuenta que lo que tratamos es de estimar la varianza de la población, no de calcularla de forma exacta.<br>

En cambio, si la muestra es relativamente pequeña (digamos $n<30$) entonces la varianza muestral de la muestra aproxima significativamente mejor la varianza de la población que la varianza.<br>

La diferencia entre desviación típica y desviación típica muestral es análoga.

Con `R`, calcularemos la varianza y la desviación típica **muestrales**. Con lo cual, si queremos calcular las que no son muestrales, tendremos que multiplicarlas por $\frac{n-1}{n}$, donde $n$ es el tamaño de la muestra. Lo veremos a continuación.<br><br>

### **Varianza y desviación típica**<br><br>

Nótese que tanto la varianza como la desviación típica dan una información equivalente. Entonces, es comprensible preguntarse por qué se definen ambas medidas si con una basta. Pues bien, las unidades de la varianza (metros, litros, años...), ya sea muestral o no, están al cuadrado, mientras que las de la desviación típica no.


## **Medidas de forma** <a name="forma"></a>

El **coeficiente de asimetría** o **sesgo** se refiere a la asimetría de la función. Cuando se acerca a cero la función tiende a ser simétrica. Si el sesgo se extiende a la derecha es negativo, por el contrario, si el sesgo se extiende hacia la izquierda es positivo.<br><br>

$$\mbox{Sesgo}= \frac{\sum(x_i - \overline{x})^3\cdot f_i}{s^3}$$<br><br>

El **coeficiente de curtosis** mide que tan puntiaguda es la función. Es mesocúrtica si es cercana a cero, similar a la distribución normal, platicúrtica (más plana con colas ligeras) si es negativa y leptocúrtica (más puntiaguda con colas pesadas) si es positiva.<br><br>

$$\mbox{Curtosis}= \frac{\sum(x_i - \overline{x})^4\cdot f_i}{s^4} - 3 $$

## **Diagramas de caja** <a name="caja"></a>

El conocido **diagrama de caja** o **box plot** es un tipo de gráfico que básicamente, remarca 5 valores estadísticos:<br><br>

- La mediana, representada por la línea gruesa que divide la caja
- El primer y tercer cuartil, que son los lados inferior y superior, respectivamente. De este modo, la altura de la caja es el rango intercuantílico
- Los extremos, los valores $b_{inf},b_{sup}$, son los <l class = "definition">bigotes</l> (<l class = "definition">whiskers</l>) del gráfico. Si $m$ y $M$ son el mínimo y máximo de la variable cuantitativa, entonces los extremos se calculan del siguiente modo:<br><br>

$$b_{inf}=\max\{m,Q_{0.25}-1.5(Q_{0.75}-Q_{0.25})\}$$
$$b_{sup}=\min\{M,Q_{0.75}+1.5(Q_{0.75}-Q_{0.25})\}$$<br><br>

- <l class = "definition">Valores atípicos</l> o <l class = "definition">outliers</l>, que son los que están más allá de los bigotes. Se marcan como puntos aislados.<br><br>

### **Más sobre los bigotes**<br><br>

Por su definición, concluimos que los bigotes marcan el mínimo y máximo de la variable cuantitativa, a no ser que haya datos muy alejados de la caja intercuantílica.

En tal caso, el bigote inferior marca el valor 1.5 veces el rango intercuantílico por debajo de $Q_{0.25}$, mientras que el superior marca el valor 1.5 veces el rango intercuantílico por encima de $Q_{0.75}$

## **Datos agrupados** <a name="agrupados"></a>

Aunque no seamos completamente conscientes de ello, tendemos a agrupar datos cuantitativos constantemente. Decimos que una persona tiene 20 años cuando se encuentra en el intervalo [20,21). Es decir, cuando ha cumplido los 20 pero aún no tiene los 21.

En estadística, existen innumerables motivos por los cuales nos interesa agrupar los datos cuando estos son cuantitativos. Uno de estos motivos puede ser perfectamente que los datos sean muy heterogéneos. En este caso, nos encontraríamos con que las frecuencias de los valores individuales serían todas muy similares, lo que daría lugar a un diagrama de barras muy difícil de interpretar.

Otro de los motivos por el que necesitamos muchas veces agrupar los datos cuantitativos es porque la precisión infinita no existe. Por tanto, esta imposibilidad de medir de manera exacta muchas de las magnitudes continuas (tiempo, peso, altura...) nos obliga a trabajar con aproximaciones o redondeos de valores reales y que cada uno de estos represente todo un intervalo de posibles valores.<br><br>


Por lo general, existen 3 situaciones en las cuales conviene sin lugar a dudas agrupar datos cuantitativos en intervalos, también llamados **clases**:<br><br>

- Cuando los datos son continuos, su redondeo ya define un agrupamiento debido a la inexistencia de precisión infinita
- Cuando los datos son discretos, pero con un número considerablemente grande de posibles valores
- Cuando tenemos muchísimos datos y estamos interesados en estudiar las frecuencias de sus valores<br><br>

### **Cómo agrupar datos**<br><br>

Antes de estudiar unos datos agrupados, hay que, obviamente, agruparlos. Este proceso consta de 4 pasos:

1. Decidir el número de intervalos que vamos a utilizar
2. Decidir la amplitud de estos intervalos
3. Acumular los extremos de los intervalos
4. Calcular el valor representativo de cada intervalo, su **marca de clase**

No hay una forma de agrupar datos mejor que otra. Eso sí, cada uno de los diferentes agrupamientos para un conjunto de datos podría sacar a la luz características diferentes del conjunto.<br><br>

### **Estableciendo el número de clases**<br><br>

En este tema explicaremos una receta para agrupar datos. Lo dicho, ni mejor ni peor que el resto.

Lo primero es establecer el número $k$ de clases en las que vamos a dividir nuestros datos. Podemos decidir en función de nuestros intereses o podemos hacer uso de alguna de las reglas existentes. Destacaremos las más populares. Sea $n$ el número total de datos de la muestra<br><br>

- **Regla de la raíz cuadrada**: $k = \lceil\sqrt{n}\ \rceil$

- **Regla de Sturges**: $k = \lceil 1+\log_{2}(n)\rceil$

- **Regla de Scott**: Se determina primero la **amplitud teórica**, $A_S$ de las clases<br><br>
 $$A_S = 3.49\cdot\tilde{s}\cdot n^{-\frac{1}{3}}$$<br>

  donde $\tilde{s}$ es la desviación típica muestral. Luego se toma<br><br>
$$k = \left\lceil \frac{\max(x)-\min(x)}{A_S}\right\rceil$$<br><br>

- **Regla de Freedman-Diaconis**: Se determina primero la **amplitud teórica**, $A_{FD}$ de las clases<br><br>

 $$A_{FD} = 2\cdot(Q_{0.75}-Q_{0.25})\cdot n^{-\frac{1}{3}}$$<br><br>
 (donde, recordemos, $Q_{0.75}-Q_{0.25}$, es el rango intercuantílico) y entonces <br><br>

$$k = \left\lceil \frac{\max(x)-\min(x)}{A_{FD}}\right\rceil$$<br><br>

- **Regla de Rice**: $k = \lceil 2 \sqrt[3]{n}\ \rceil$<br><br>


Las dos primeras solo dependen de $n$, mientras que las dos últimas también tienen en cuenta, de formas diferentes, la dispersión de los datos. De nuevo, no hay ninguna mejor que las demás. Pero sí puede ocurrir que métodos diferentes den lugar a la observación de características diferentes en los datos.<br><br>

### Estableciendo el número de clases con R<br><br>

<br><br>

Puede ocurrir que las difrentes reglas den valores diferentes, o no.<br><br>

### **Decidiendo la amplitud**<br><br>

Una vez determinado $k$, hay que decidir su amplitud.

La forma más fácil y la que nosotros utilizaremos por defecto es que la amplitud de todos los intervalos sea la misma, $A$. Esta forma no es la única.

Para calcular $A$, lo que haremos será dividir el rango de los datos entre $k$, el número de clases, y redondearemos por exceso a un valor de la precisión de la medida.

Si se da el improbable caso en que el cociente de exacto, tomaremos como $A$ ese cociente más una unidad de precisión.<br><br>

### **Extremos de los intervalos**<br><br>

Es la hora de calcular los extremos de los intervalos. Nosotros tomaremos estos intervalos siempre cerrados por su izquierda y abiertos por la derecha, debido a que esta es la forma en que R los construye y porque es así como se utilizan en Teoría de Probabilidades al definir la distribución de una variable aleatoria discreta y también en otras muchas situaciones cotidianas.<br><br>

Utilizaremos la siguiente notación<br><br>

$$[L_1,L_2),[L_2,L_3),\dots,[L_k,L_{k+1})$$<br><br>

donde los $L_i$ denotan los extremos de los intervalos. Estos se calculan de la siguiente forma:<br><br>

$$L_1 = \min(x)-\frac{1}{2}\cdot \text{precisión}$$<br><br>

A partir de $L_1$, el resto de intervalos se obtiene de forma recursiva:<br><br>

$$L_2 = L_1 + A$$
$$L_3 = L_2 + A$$
$$\vdots$$
$$L_{k+1} = L_k+A$$<br><br>

Si nos fijamos bien, los extremos forman una progresión aritmética de salto $A$:<br><br>

 $$L_{i} = L_{1}+(i-1)A,\qquad i=2,\dots,k+1$$<br><br>

De esta forma garantizamos que los extremos de los intervalos nunca coincidan con valores del conjunto de datos, puesto que tinen una precisión mayor.<br><br>


### **Marca de clase**<br><br>

Solo nos queda determinar la **marca de clase**, $X_i$, de cada intervalo $[L_i,L_{i+1})$.

Este no es más que un valor del intervalo que utilizaremos para identificar la clase y para calcular algunos estadísticos.Genralmente,<br><br>

 $$X_i = \frac{L_i+L_{i+1}}{2}$$<br><br>

es decir, $X_i$ será el punto medio del intervalo, para así garantizar que el error máximo cometido al describir cualquier elemento del intervalo por medio de su marca de clase sea mínimo o igual a la mitad de la amplitud del respectivo intervalo.


Es sencillo concluir que, al tener todos los intervalos amplitud $A$, la distancia entre $X_i$ y $X_{i+1}$ tambien será $A$. Por consiguiente,<br><br>

$$X_{i} = X_1+ (i-1)A,\qquad i=2,\dots,k$$<br><br>

donde<br>

$$X_1 = \frac{L_1+L_2}{2}$$

## **Estudiando datos agrupados** <a name="estudio_grupos"></a>

### Frecuencias<br>

Una primera consideración es tratar las clases obtenidas en el paso anterior como los niveles de una variable ordinal y calcular sus frecuencias.

- La frecuencia absoluta de una clase será el número de datos originales que pertenecen a la clase
- La frecuencia absoluta acumulada de una clase será el número de datos que pertenecen a dicha clase o alguna de las anteriores<br><br>

### Tabla de frecuencias<br><br>

Normalmente, las frecuencias de un conjunto de datos agrupados se suele representar de la siguiente forma<br><br><center>


| Intervalos | $X_j$ | $n_j$ | $N_j$ | $f_j$ | $F_j$ |  
|-------------------|--------------------|--------------------|--------------------|--------------------|--------------------|
|$[L_1,L_2)$|$X_1$|$n_1$|$N_1$|$f_1$|$F_1$|
|$[L_2,L_3)$|$X_2$|$n_2$|$N_2$|$f_2$|$F_2$|
|$\vdots$|$\vdots$|$\vdots$|$\vdots$|$\vdots$|$\vdots$|
|$[L_k,L_{k+1})$|$X_k$|$n_k$|$N_k$|$f_k$|$F_k$|

</center>

## **Estadísticos para datos agrupados** <a name="estadisticos"></a>

Al tener una muestra de datos numéricos, conviene calcular los **estadísticos** antes de realizar los agrupamientos, puesto que de lo contrario podemos perder información.<br><br>

No obstante, hay situaciones en que los datos los obtenemos ya agrupados. En estos casos, aún sigue siendo posible calcular los estadísticos y utilizarlos como aproximaciones de los estadísticos de los datos "reales", los cuales no conocemos.<br><br>

La media $\bar{x}$, la varianza, $s^2$, la varianza muestral, $\tilde{s}^2$, la desviación típica, $s$, y la desviación típica muestral, $\tilde{s}$ de un conjunto de datos agrupados se calculan mediante las mismas fórmulas que para los datos no agrupados con la única diferencia de que sustituimos cada clase por su marca de clase y la contamos con su frecuencia.<br><br>

Es decir, si tenemos $k$ clases, con sus respectivas marcas $X_1,\dots,X_k$ con frecuencias absolutas $n_1,\dots,n_k$ de forma que $n=\sum_{j=1}^kn_j$. Entonces<br><br>

$$\bar{x}=\frac{\sum_{j=1}^kn_jX_j}{n}, \mu=\frac{\sum_{j=1}^kn_jX_j}{N}$$

$$\quad \sigma^2=\frac{\sum_{j=1}^kn_jX_j^2}{N}-\mu^2  = \quad \sigma^2=\frac{\sum_{j=1}^k n_j(X_j-\mu)^2}{N}, \quad s^2=\frac{\sum_{j=1}^k n_j(X_j-\bar{x})^2}{n-1} = \quad \tilde{s}^2=\frac{n}{n-1}\cdot \sigma^2$$
$$\sigma=\sqrt{\sigma^2},\quad \tilde{s}=\sqrt{\tilde{s}^2}$$
$$N=\sum_{j=1}^k n_j,  n=\sum_{j=1}^k n_j $$
<br><br>

### **Intervalo modal**<br><br>

En lo referente a la moda, esta se sustituye por el **intervalo modal**, que es la clase con mayor frecuencia (absoluta o relativa, tanto da).

En el caso en que un valor numérico fuera necesario, se tomaría su marca de clase.<br><br>

### **Intervalo crítico para la mediana**<br><br>

Se conoce como **intervalo crítico para la mediana**, $[L_c,L_{c+1})$, al primer intervalo donde la frecuencia relativa acumulada sea mayor o igual que 0.5

Denotemos por $n_c$ su frecuencia absoluta, por $A_c = L_{c+1}-L_c$ su amplitud y por $N_{c-1}$ la frecuencia acumalada del intervalo inmediantamente anterior (en caso de ser $[L_c,L_{c+1})=[L_1,L_2)$, entonces $N_{c-1}=0$). Entonces, $M$ será una aproximación para la mediana de los datos "reales" a partir de los agrupados<br><br>

$$M = L_c +A_c\cdot\frac{\frac{n}{2}-N_{c-1}}{n_c}$$
<br><br>

### **Aproximación de los cuantiles**<br><br>

La fórmula anterior nos permite aproximar el cuantil $Q_p$ de los datos "reales" a partir de los datos agrupados:<br><br>

$$Q_p = L_p +A_p\cdot\frac{p\cdot n-N_{p-1}}{n_p}$$<br><br>

donde el intervalo $[L_p,L_{p+1})$ denota el primer intervalo cuya frecuencia relativa acumulada es mayor o igual a $p$

```python
def percentile(data, percentile):
  int_critico = data[data['Frecuencia relativa acumulada'] >= percentile].index
  n = data["Frecuencia absoluta acumulada"].max()
  #ind = data.index.get_loc(int_critico[0]) uso de at
  Linf = int_critico[0].left
  Lsup = int_critico[0].right
  A = Lsup - Linf
  Nc = data.iloc[data.index.get_loc(int_critico[0]) - 1]['Frecuencia absoluta acumulada']
  nc = data.loc[int_critico[0]]['Frecuencia absoluta']
  per = Linf + A * ((n * percentile) - Nc) / nc
  return per
```

```python
def percentiles(data, percentile):
  int_critico = data[data['Frecuencia relativa acumulada'] >= percentile].index
  Nc = data.iloc[data.index.get_loc(int_critico[0]) - 1]['Frecuencia absoluta acumulada']
  nc = data.loc[int_critico[0]]['Frecuencia absoluta']
  return int_critico[0].left + (int_critico[0].right - int_critico[0].left) * ((data["Frecuencia absoluta acumulada"].max() * percentile) - Nc) / nc
```

## **Histogramas** <a name="histogramas"></a>

La mejor manera de representar datos agrupados es mediante unos diagramas de barras especiales conocidos como **histogramas**.</p>

<p>En ellos se dibuja sobre cada clase una barra cuya área representa su frecuencia. Podéis comprobar que el producto de la base por la altura de cada barra es igual a la frecuencia de la clase correspondiente.</p><br>

### **El uso de histogramas**<br><br>

<p>Si todas las clases tienen la misma amplitud, las alturas de estas barras son proporcionales a las frecuencias de sus clases, con lo cual podemos marcar sin ningún problema las frecuencias sobre el eje vertical. Pero si las amplitudes de las clases no son iguales, las alturas de las barras en un histograma no representan correctamente las frecuencias de las clases.</p>

<p>En este último caso, las alturas de las barras son las necesarias para que el área de cada barra sea igual a la frecuencia de la clase correspondiente y como las bases son de amplitudes diferentes, estas alturas no son proporcionales a las frecuencias de las clases, por lo que no tiene sentido marcar las frecuencias en el eje vertical</p>

<p>Los histogramas también son utilizados para representar frecuencias acumuladas de datos agrupados. En este caso, las alturas representan las frecuencias independientemente de la base debido a que éstas deben ir creciendo.</p><br><br>

### **Interpretación de los histogramas** <br><br>

<p>El eje de las abcisas representa los datos. Aquí marcamos los extremos de las clases y se dibuja una barra sobre cada una de ellas. Esta barra tiene significados diferentes en función del tipo de histograma, pero en general representa la frecuencia de su clase</p><br>

<ul>
<li>Histograma de frecuencias absolutas: la altura de cada barra es la necesaria para que el área de la barra sea igual a la frecuencia absoluta de la clase. Las amplitudes de las clases pueden ser todas iguales o no. En el primer caso, las alturas son proporcionales a las frecuencias. En el segundo caso, no existe tal proporcionalidad. De todas formas, sea cual sea el caso, conviene indicar de alguna forma la frecuencia que representa cada barra.</li>
</ul><br>


<p></p><br>


<ul>
<li>Histograma de frecuencias relativas: la altura, densidad, de cada barra es la necesaria para que el área sea igual a la frecuencia relativa de la clase. La suma de todas las áreas debe ser 1. De nuevo, conviene indicar de alguna forma la frecuencia que representa cada barra.</li><br>

<li>Histogramas de frecuencias acumuladas: las alturas de las barras son iguales a las frecuencias acumuladas de las clases, independientemente de su amplitud.</li>
</ul><br>

### **Frecuencias nulas** <br><br>

<p>No es conveniente que en un histograma aparezcan clases con frecuencia nula, exceptuando el caso en que represente poblaciones muy diferentes y separadas sin individuos intermedios.</p>

<p>Si apareciesen clases vacías, convendría utilizar un número menor de clases, o bien unir las clases vacías con alguna de sus adyacentes. De este último modo romperíamos nuestro modo de trabajar con clases de la misma amplitud.</p>

### **Gráficos de densidad**

#### **Histogramas de frecuencias relativas** <br><br>

En estos histogramas, es común superponer una curva que estime la densidad de la distribución de la variable cuantitativa definida por la característica que estamos midiendo.<br><br>

La **densidad** de una variable es una curva cuya área comprendida entre el eje de las abcisas y la propia curva sobre un intervalo es igual a la fracción de individuos de la población que caen dentro de ese intervalo.<br><br>

Para hacernos una idea visual, imagina que va aumentando el tamaño de la muestra a la vez que agrupas los datos en un conjunto cada vez mayor de clases. Si el rango de los datos se mantiene constante, la amplitud de las clases del histograma irá menguando. Además, cuando \(n\), el tamaño de la muestra, tiende a infinito, los intervalos tienden a ser puntos y, a su vez, las barras tienden a ser líneas verticales. Pues bien, los extremos superiores de estas líneas serán los que dibujen la densidad de la variable.<br><br>

#### **Campana de Gauss**<br><br>

Es la densidad más famosa: la **Campana de Gauss** se corresponde con una variable que siga una distribución nomal.<br><br>

La forma de la campana depende de dos parámetros: el valor medio, $(\mu)$, y su desviación típica, $(\sigma)$.<br><br>

<center>

#### **Histogramas de frecuencias relativas acumuladas**<br><br>

En este último tipo de histograma, se suele superponer una curva que estime la **función de distribución** de la variable definida por la característica que estamos midiendo.<br><br>

Esta función de distribución, en cada punto nos da la fracción de individuos de la población que caen a la izquierda de este punto: su frecuencia relativa acumulada.<br><br>

En general, la función de distribución en un valor determinado se obtiene hallando el área de la función de densidad que hay a la izquierda del valor.
